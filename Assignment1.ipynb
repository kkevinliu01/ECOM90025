```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
```

**Question 1**

Read the training sample to a Pandas dataframe.

```python
url_train = "https://raw.githubusercontent.com/kkevinliu01/ECOM90025-A1/main/train.csv"
df_train = pd.read_csv(url_train, index_col=0)
```

```python
df_train.head(5)
```

**Question 2**

Report the correlations of explanatory variables with the dependent variable $y$.

```python
corr = df_train.corr(method = 'pearson')
```

```python
corr = corr.drop(corr.columns[0], axis=1).head(1).T #remove correlation of y on y and take only correlations of x on y
corr
```

**Question 3**

Choose one explanatory variable, denoted by $y = \beta_0 + \beta_1 x + \epsilon$ , and estimate a simple linear model 

```python
corr_abs = abs(corr)

x_chosen = corr_abs.sort_values(by=['Y'], ascending=False).head(1).index.values[0] #find highest correlation variable

x_chosen
```

```python
df_estimate = df_train[["Y", x_chosen]]

df_estimate.head(5)
```

```python
# estimate the linear model

length = df_estimate.shape[0]
X = df_estimate[x_chosen].values.reshape(length,1)
y = df_estimate["Y"].values.reshape(length,1)

reg = LinearRegression().fit(X, y)

b0 = reg.intercept_[0]

b1 = reg.coef_[0][0]
```

**Question 4**

Use the estimtes of $\beta_0$ and $\beta_1$ and the testing sample to predict $y$.

```python
url_test = "https://raw.githubusercontent.com/kkevinliu01/ECOM90025-A1/main/test.csv"
df_test = pd.read_csv(url_test, index_col=0)

length = df_test.shape[0]

df_test = df_test[x_chosen].values.reshape(length, 1)

predictions = reg.predict(df_test)
```

```python
url_submission = "https://raw.githubusercontent.com/kkevinliu01/ECOM90025-A1/main/sample_submission.csv"
df_submission = pd.read_csv(url_submission, index_col=0)

df_submission["Y"] = predictions

df_submission.head(5)
```
